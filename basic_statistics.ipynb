```python
#Libraries
import pandas as pd
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("Baseball.csv")
df.head()
```

#Mean

```python
#mean of Runs scored
df['RS'].mean()
```

```python
#mean of Runs scored by the ARI team
df.loc[df['Team']=='ARI'].RS.mean()
```

```python
#mean of Runs scored by the ARI team since 2005
df.loc[(df.Team == 'ARI') & (df.Year >= 2005)]['RS'].mean()
```

```python
#question: what is the mean of runs allowed (RA) of Chicago(CHC) until 2007
df.loc[(df.Team == 'CHC') & (df.Year <= 2007)].RA.mean()
```

#Medium

```python
#Mean and medium of wins (W)
print(df.W.mean())
df.W.median()
```

```python
#question: what is the medium of wins of the Baltimore team (BAL) until 2000
df.loc[(df.Team == 'BAL') & (df.Year < 2000)].W.median()
```

#Mode

```python
# Mode, Medium, Mean of OBP
print(df.OBP.mean())
print(df.OBP.median())
print(df.OBP.mode())
```

```python
#Question: mode of OBP during the year 2010
df.loc[df.Year == 2010].OBP.mode()
```

#Correlation

```python
#pick variables
df_correlation = df[['RS', 'RA', 'W']]
df_correlation.head(1)
```

```python
df_correlation.corr()
```

```python
#correlation heatmap
sns.heatmap(data = df_correlation.corr(),
            annot = True,
            fmt = '.2g',
            center = 0,
            cmap = 'coolwarm',
            linewidths = 1,
            linecolor = 'black')
```

```python
#challenge: correlation matrix between OBP, SLG and BA. 
# heatmap with 2 parameter change
df_corr = df[['OBP', 'SLG', 'BA']]
df_corr.corr()
```

```python
sns.heatmap(data = df_corr.corr(),
            center = 0.7,
            annot = True,
            fmt = '.2g',
            cmap = 'rocket',
            linewidths=1,
            linecolor='orange'
            )
```

#standard deviation

```python
print(df.OOBP.mean())
print(df.OOBP.std()) #0.15 means the data is close to each other
```

```python
#Question: what is the standard deviation of BA
print(df.BA.mean())
print(df.BA.std())
```
